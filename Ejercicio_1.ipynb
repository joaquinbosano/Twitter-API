```python

class StdOutListener(StreamListener):
    """
    Printea tweets recibidos a output estandar.
    """
    def __init__(self, fetched_tweets_filename):
        
        self.fetched_tweets_filename = fetched_tweets_filename

    def on_data(self, data):
        try:
            print(data)
            
            with open(self.fetched_tweets_filename, 'a') as tf:
                tf.write(data)
            return True
        
        except BaseException as e:
            print("Error on_data %s" % str(e))
        
        return True

```

```python

class TwitterStreamer():
    """
    Streaming tweets en tiempo real.
    """
    def __init__(self):
        pass

    def stream_tweets(self, fetched_tweets_filename, hash_tag_list):
        
        # Autenticacion y conexion con nuestra clase
        listener = StdOutListener(fetched_tweets_filename)
        stream = Stream(auth, listener)

        # Filtra el Stream para capturar solo el hashtaf:
        stream.filter(track=hash_tag_list)
```

```python
# Setear hastags y nombre de archivo.

hash_tag_list = ["donald trump"]
fetched_tweets_filename = "tweets.txt"



twitter_streamer = TwitterStreamer()
```

```python
# stream_tweets


twitter_streamer.stream_tweets(fetched_tweets_filename, hash_tag_list)
```
